# Exploratory Data Analysis for Customer Churn Prediction
Welcome to the Exploratory Data Analysis of our telecom customer data. This analysis is crucial for understanding the factors that lead to customer churn. By identifying these factors, we can develop strategies to improve customer retention, optimize service offerings, and ultimately enhance profitability.

## Introduction
This EDA provides a comprehensive view of the dataset involving customer demographics, service usage, billing information, and churn status. Our goal is to identify patterns and insights that could help predict and reduce churn rates.

## Dataset Overview
- **Customer ID**: A unique identifier for the customer.
- **Gender**: The customer's gender (Male, Female, Non-binary).
- **Age**: The customer's age.
- **Tenure Months**: Number of months the customer has been with the company.
- **Phone Service**: Indicates whether the customer has phone service (Yes, No).
- **Multiple Lines**: Indicates whether the customer has multiple lines (Yes, No, No phone service).
- **Internet Service**: Type of internet service (DSL, Fiber optic, None).
- **Online Security**: Indicates whether the customer subscribes to an additional online security service (Yes, No, No internet service).
- **Online Backup**: Indicates whether the customer subscribes to an additional online backup service (Yes, No, No internet service).
- **Device Protection**: Indicates whether the customer subscribes to an additional device protection plan (Yes, No, No internet service).
- **Tech Support**: Indicates whether the customer subscribes to additional tech support (Yes, No, No internet service).
- **Contract**: Type of contract the customer has (Month-to-month, One year, Two year).
- **Monthly Charges**: The amount charged to the customer each month.
- **Total Charges**: The total amount charged to the customer.
- **Churn**: Whether the customer left within the last month (Yes, No).

# Data Analysis and Visualization
## Univariate Analysis
Distribution of each variable (Histograms, Pie Charts)

Statistics summary

Gender vs Churn

Exploring if gender plays a role in churn.

```python
sns.countplot(x='gender', hue='churn', data=df)
plt.title('Churn Rate by Gender')
plt.show()
```

### Age and Tenure Distribution
Understanding how age and tenure might influence churn.

```python
sns.histplot(df['age'])
plt.title('Distribution of Age')
plt.show()

```

```python
sns.histplot(df['tenure_months'])
plt.title('Distribution of Tenure Months')
plt.show()
```

### Services Subscribed vs Churn
Do specific services like online security, tech support impact churn rates?

```python
services = ['online_security', 'online_backup', 'device_protection', 'tech_support']

for service in services:
    sns.countplot(x=service, hue='churn', data=df)
    plt.title(f'Churn Rate by {service.title()}')
    plt.show()
```

### Churn by Monthly and Total Charges

```python
sns.kdeplot(data=df, x='monthly_charges', hue='churn', fill=True)
plt.title('Density Plot of Monthly Charges by Churn')
plt.show()
```

```python
sns.kdeplot(data=df, x='total_charges', hue='churn', fill=True)
plt.title('Density Plot of Total Charges by Churn')
plt.show()
```

### Correlation Analysis
Heatmap of correlations between features.

```python
# Drop non-numeric columns
df_numeric = df.drop(columns=['customer_id', 'gender', 'phone_service', 'multiple_lines', 'internet_service', 'online_security', 'online_backup', 'device_protection', 'tech_support', 'contract', 'churn'])

# Compute correlation matrix
corr_matrix = df_numeric.corr()

# Visualize correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
