# Key Aspects of Exploratory Data Analysis
1. Descriptive Statistics
   - Summary Statistics - Measure of central Tendancy - Mean, mode, median
   - Measure of dispersion - Range, Starndard deviation, percentiles
   - Data Distribution - Histrogram, Boxplot, Probability Density Plots

2. Data Visualization
  - Scatter plot - Identifying the relationship of 2 variables so we can understand the trends, patterns, and correlation
  - Pairplots - Multivariate analysis
  - line plots - Time series analysis, observe trends, patterns over time
  - Bar plots - Visualize the categorical variables in the data

3. Handling Missing Values
  - Identify the missing values - Assess the presence of missing
  - Impute missing values - remove, impute -mean, mode, mediate, ...

4. Outliers
  - Boxplot - Identify the outliers in the numerical data
  - scatter plots
  - Z-Score

5. Feature Engineering
  - create new features - Features having more information on the analysis
  - Transformation of the variables - Log transformation

6. Correlation Analysis
  - Correlation Matrix - To understand the relationships
  - heatmaps - Visualize the matrices for a better understanding

7. Understanding the relationships
  - grouping and aggregations
  - cross tabulation

8. Statistical Testing
  - Hypothesis test

9. Interactive Exploration
  - Interactive Dashboads, - Plotly, tableau, streamlit
  - Widgets, and filters

10. Documentation and communication
  - Document and create presentation - reports, presentation, visualizations

# Groupby

```python
# Data
data = {"Company": ["Google", 'Google', 'Microsoft', 'Microsoft', 'Facebook', 'Facebook'],
        'Person': ['Sam', 'Chalie', 'Amy', 'Carl', 'Sara', 'Venesa'],
        'Sales': [200, 120, 340, 124, 234, 350]}
```

```python
# Convert the data to a dataframe
df = pd.DataFrame(data)
```

```python
df.drop('Person', axis = 1, inplace = True)
```

```python
df
```

```python
# Group by the companies
group = df.groupby('Company')
```
